```python
# Packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
```

```python
# Use LaTeX font
plt.rcParams.update({'text.usetex': True})

# Figure font config
label_font = {'fontfamily': 'Arial Black', 'fontsize': 14}
title_font = {'fontfamily': 'Arial Black', 'fontsize': 16}
legend_font = {'family': 'Palatino Linotype', 'size': 12}
text_font = {'family': 'Palatino Linotype', 'size': 12}
```

```python
# Load data
particle = []
velocity = []
for i in range(6):
    particle.append(np.load('../data/multi-matshow/particle_' + str(i) + '.npy'))
    velocity.append(np.load('../data/multi-matshow/velocity_' + str(i) + '.npy'))
```

```python
# Plot
fig, axs = plt.subplots(1, 6, figsize=(16, 3))

for i in range(6):
    for idx, (x, y) in enumerate(zip(particle[i][:, 0], particle[i][:, 1])):
        axs[i].scatter(
            x, y, 
            marker='o',
            color=cm.RdYlBu_r(velocity[i][idx])
        )
    
    # Label and title
    axs[i].get_xaxis().set_visible(False)
    axs[i].get_yaxis().set_visible(False)
    axs[i].set_title(f'$t={i}$', fontdict=title_font)

    # Axis range
    axs[i].set_xlim([0.5, 5.5])
    axs[i].set_ylim([0.5, 4])

plt.tight_layout()
plt.savefig('../fig/multi-matshow.jpg', dpi=300)
```
